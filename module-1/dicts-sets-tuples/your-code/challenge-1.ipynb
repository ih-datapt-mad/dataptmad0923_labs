#### How many times does each letter in `letters` appear in `tup3`?

Print out the number of occurrence of each letter.

````python
# Your code here
letters = ["a", "b", "c", "d", "e"]
for letter in letters:
    if letter in tup3:
        count = tup3.count(letter)
        print(count)
    else:
        print(0)
````
