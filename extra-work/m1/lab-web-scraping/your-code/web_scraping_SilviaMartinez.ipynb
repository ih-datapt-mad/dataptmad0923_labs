#### Download, parse (using BeautifulSoup), and print the content from the Trending Developers page from GitHub:

````python
parsed_html.body.a
````

````python
#Find 1 "h1" tag:

article = parsed_html.find("h1")
article
````

````python
#Find all "h1" tags que además tenga otra subclase llamada h3...:

article = parsed_html.find_all("h1", {"class" : "h3 lh-condensed"})
article
````

````python
article[0].contents
````

````python
article[0].contents[1].string
````

````python
#Creo una lista vacía. Article era una lista con 25 etiquetas que había. Para cada elemento en article, voy a guardar una string. Luego lo parseo con bs4 y encuentro todas las "a" que tienen eso. Para cada elemento de la lista que me devuelve, hago append a mi lista en formato de string y con un strip que quita espacios en blanco.

articles = []
for a in article:
    a = str(a)
    soup_a = bs4.BeautifulSoup(a, "html.parser")
    list_a = soup_a.find_all("a")
    for e in list_a:
        articles.append(e.string.strip())
print(len(articles))
````

````python
articles
````

####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
response = requests.get(url)
````

````python
response.status_code
````

````python
html = response.content
````

````python
parsed_html = bs4.BeautifulSoup(html, "html.parser")
````

````python
parsed_html
````

````python
eq_table = parsed_html.find("table", {"class" : "eqs"})
eq_table
````

````python
if eq_table:
    eq_rows = eq_table.find("tbody").find_all("tr")
    dates = []
    latitudes = []
    longitudes = []
    depths = []
    magnitudes = []
    regions = []

    for row in eq_rows:
        columns = row.find_all("td")
        date_time = columns[0].text.strip()
        latitude = columns[1].text.strip()
        longitude = columns[2].text.strip()
        depth = columns[3].text.strip()
        magnitude = columns[4].text.strip()
        region = columns[5].text.strip()
        
        dates.append(date_time)
        latitudes.append(latitude)
        longitudes.append(longitude)
        depths.append(depth)
        magnitudes.append(magnitude)
        regions.append(region)
````

````python
eq_df = pd.DataFrame({
            "Date & Time": dates,
            "Latitude": latitudes,
            "Longitude": longitudes,
            "Depth": depths,
            "Magnitude": magnitudes,
            "Region": regions
        })
````

````python
eq_df
````

#### Find the book name, price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
response = requests.get(url)
response.status_code
````

````python
html = response.content
````

````python
parsed_html = bs4.BeautifulSoup(html, "html.parser")
````

````python
item = parsed_html.find_all("article", {"class" : "product_pod"})
item
````

````python
title = parsed_html.find_all("a", {"href" : True, "title": True})
title
````

````python
titles = []
for t in title:
    titles.append(t.text)
titles
````

````python
price = parsed_html.find_all("p", {"class" : "price_color"})
price
````

````python
prices = []
for p in price:
    prices.append(p.text)
prices
````

````python
avail = parsed_html.find_all("p", {"class" : "instock availability"})
avail
````

````python
avails = []
for a in avail:
    avails.append(a.text.replace("\n", "").replace("  ", ""))
avails
````

````python
df = pd.DataFrame()
````

````python
df["Title"] = titles
df["Price"] = prices
df["Availability"] = avails
df.index = np.arange(1, len(df)+1)
df
````
