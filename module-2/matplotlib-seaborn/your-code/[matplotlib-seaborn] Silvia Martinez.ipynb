#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

fig, [ax1, ax2] = plt.subplots(1,2, figsize = [10,10])
y = x**2
ax1.plot(x, y)
ax1.set_title("Normal Scale")
ax1.set_xlabel("x label")
ax1.set_ylabel("y label")
ax1.set_xscale("linear")

y = np.exp(x) #no me deja representarlo como "e"
ax2.plot(x, y)
ax2.set_title("Logaritmic Scale")
ax2.set_xlabel("x label")
ax2.set_ylabel("y label")
ax2.set_yscale("log")

fig.suptitle("Normal and logaritmic scales", fontsize=16, fontweight = "bold")

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize = [12,4])

median_month = fitbit[["Steps", "Months_encoded"]].groupby("Months_encoded").agg("median")
ax1.set_xticks(range(12))
ax1.set_title("Steps vs. Months_encoded")
median_month.plot.barh(ax=ax1)

median_work_or_weekend = fitbit[["Steps", "Work_or_Weekend"]].groupby("Work_or_Weekend").agg("median")
ax2.set_title("Steps vs. Work_or_Weekend")
median_work_or_weekend.plot.barh(ax=ax2)


plt.savefig("../figures/result.png")
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib

fig, ax = plt.subplots(figsize=(9, 9))

x = titanic["Age"]
y = titanic["Fare"]

ax.scatter(x, y, s=60, alpha=0.7, edgecolors="k")

b, a = np.polyfit(x, y, deg=1)
xseq = np.linspace(x.min(), x.max(), num=100)
ax.plot(xseq, a + b * xseq, color="k", lw=2.5);

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Linear Relationship between Age and Fare")

plt.show()
```

```python
# Method 2 - seaborn

sns.lmplot(x="Age",y="Fare", data = titanic, order=1, line_kws={'color': 'red'})
plt.title("Linear Relationship between Fare and Age")
```
